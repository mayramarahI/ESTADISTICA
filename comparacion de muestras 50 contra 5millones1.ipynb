## Con una semilla inicial de 555 se van a comparar los resumenes de la muestra1 con 50 aleatorios y la muestra2 con 5 millones de aleatorios, con una distribución normal para ambas muestras (Teorema de Límite Central).

```r
set.seed(555)
muestra1 <- rnorm(50,10,0.5)
summary(muestra1)
```

```r
muestra2 <- rnorm(5e6,10,0.5)
summary(muestra2)
```

## Como se observa la media para la muestra1 y muestra2 son casi iguales. Así como los valores del primer y tercer cuartil y la mediana.

```r
layout(matrix(1:2,1,2))
```

```r
options(repr.plot.width=10,repr.plot.height=10/1.61803)
hist(muestra1,main="Histograma Muestra1",
     ylab="FRECUENCIA",xlab="Muestra1",ylim=c(0,25),xlim=c(8.3,11.5),labels=T,col="green")
##hist(muestra1)
```

```r
options(repr.plot.width=10,repr.plot.height=10/1.61803)
hist(muestra2,main="Histograma Muestra2",
     ylab="FRECUENCIA",xlab="Muestra2",ylim=c(0,8e5),xlim=c(7,13),col="green")
```

## La variación porcentual entre la media de la muestra 1 Y 2 es de 0.11%

```r
dif <- abs (mean (muestra1)-mean(muestra2)) / mean(muestra2)
```

```r
dif*100
```

## La variación porcentual entre la desviación estándar de la muestra 1 Y 2 es de 0.44%

```r
dif2 <- abs (sd (muestra1)-sd(muestra2)) / sd(muestra2)
```

```r
dif2*100
```

# De tal forma se puede concluir que de una población se puede analizar una muestra y que los parámetros de probabilidad obtenidos serán casi iguales. siempre y cuando se tenga la misma distribución normal.
