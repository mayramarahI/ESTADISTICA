## Comparación de resultados entre una población y una muestra con una distribución normal.

```r
set.seed(555)
muestra1 <- rnorm(50,10,0.5)
summary(muestra1)
```

```r
muestra2 <- rnorm(5e6,10,0.5)
summary(muestra2)
```

```r
layout(matrix(1:2,2,1))
```

```r
options(repr.plot.width=12,repr.plot.height=12/1.61803)
hist (muestra1, main="HISTOGRAMA MUESTRA1",
     col="purple", border="pink", labels=T, xlab="POBLACIÓN", ylab="FRECUENCIAS",ylim=c(0,25),xlim=c(8,11.5))
```

```r
options(repr.plot.width=8,repr.plot.height=8/1.61803)
hist (muestra2, main="HISTOGRAMA MUESTRA2",
     col="purple", border="pink", xlab="POBLACIÓN", ylab="FRECUENCIAS",ylim=c(0,8e5),xlim=c(7,13))
```

```r
dif <- abs (mean (muestra1)-mean(muestra2)) / mean(muestra2)
```

```r
dif*100
```

```r
dif2 <- abs (sd (muestra1)-sd(muestra2)) / sd(muestra2)
```

```r
dif2*100
```
